```python
from PIL import Image 

import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
# abrir imagem
image = Image.open('images/images.jpg')
image.thumbnail((128,128), Image.Resampling.LANCZOS)
#transformar os pixels da imagem em matrix
N, M = image.size
X = np.asarray(image).reshape((M*N, 3))
# mostar imagem
image
```

```python
# aplicar o k-means a estes dados
model = KMeans(n_clusters=5).fit(X)
# capturar os centros e usar como cores da paleta
cores = model.cluster_centers_.astype('uint8')[np.newaxis]
# exibir paleta de cores
plt.imshow(cores)
plt.axis('off')
plt.show()
```
